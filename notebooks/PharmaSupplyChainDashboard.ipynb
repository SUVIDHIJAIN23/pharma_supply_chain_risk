# Pharmaceutical Drug Shortage Analysis

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import datetime
from dash import Dash, dcc, html, callback, Input, Output, State
import dash_bootstrap_components as dbc
from jupyter_dash import JupyterDash
```

```python
df = pd.read_csv('../data/shortage_clean.csv')
date_cols = ['initial_posting_date', 'update_date', 'discontinued_date']
df[date_cols] = df[date_cols].apply(pd.to_datetime, errors='coerce')
categorical_columns = ['availability', 'therapeutic_category', 'dosage_form','company_name','brand_name']
availability_categories = df['availability'].unique()
therapeutic_categories = df['therapeutic_category'].unique()
dosage_categories = df['dosage_form'].unique()
```

```python
app = Dash(__name__, external_stylesheets=[dbc.themes.FLATLY])
app.layout = dbc.Container([
    html.H2("Pharmaceutical Drug Shortage Analysis"),
    ## --Row 1 for the summary
    dbc.Row([
            dbc.Col(
                    dbc.Card([html.H4("Test!")])
                    )
            ]),
            html.Br(),
    ## -- Row 2 with column 1 with all the controls and column 2 with one of the graphs
    dbc.Row([
            dbc.Col(
                    dbc.Card([
                                dbc.CardHeader("Filters"),
                                dbc.CardBody([
                                    html.Label("Categories to filter:"),
                                    dcc.Dropdown(
                                    id="categorial_dd",
                                    options=[{"label": c, "value": c} for c in categorical_columns],
                                    value="availability"                      # default selection
                                    ),
                                    html.Br(),
                                    html.Label("Select Date Range to filter:"),
                                    dcc.DatePickerRange(
                                    id="dr",
                                    start_date="2010-01-01",
                                    end_date="2026-02-01"
                                    )
                                ])
                                

                            ])





                    ,width=4), 
            dbc.Col(
                    width=8)
            ])
    ])

if __name__ =="__main__":
    app.run(jupyter_mode='external', debug = True)
```
